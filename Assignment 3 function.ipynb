```python
#6 Explain the concept of generators in Python and how they are defined.
#ANS## A generator function is a special type of function that returns an iterator object. Instead of using return to send back a single value, generator functions use yield to produce a series of results over time.
#This allows the function to generate values and pause its execution after each yield, maintaining its state between iterations.
def square_num(n):
    for i in range(n):
        yield i**2
square_num(10)
```

```python
# Geneartor Function #### Advantages.

#Lazy Evaluation: Generates values one at a time, pausing execution and resuming from where it left off.
#State Preservation: Maintains its state between yields, allowing the function to pick up where it left off on subsequent calls.
#Memory Efficiency: Does not store the entire sequence in memory, making it ideal for large or infinite sequences.
#Infinite Sequences: Can be used to produce potentially infinite sequences without consuming excessive memory.
def generate_squares(n):
    for i in range(n):
        yield i * i

```

```python
## 4##  Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    # Check if the number is less than 2, in which case it's not prime
    if n <= 1:
        return False
    # Check for factors from 2 to sqrt(n) (optimization for faster checking)
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False  # Found a divisor, so the number is not prime
    return True 
is_prime(12)
is_prime(7)
```

```python
### 5 # Create an iterator class in Python that generates the Fibonacci sequence up to a specified number ofterms.
def fib(n):
    a=0
    b=1
    for i in range (n):
        yield a
        a, b= b, a+b
fib(90)
```

```python
###6## Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(exponent):
    # Yield powers of 2 from 0 to the given exponent
    for i in range(exponent + 1):
        yield 2 ** i
powers_of_two(8)
```

```python
 ##7## Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(file_path):
    # Open the file and yield each line
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Use strip() to remove any trailing newline characters

# Example usage:
 read_file_lines ("answerofmainspaper.txt")
```

```python
 ##9# Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temps=[0, 10, 20, 30, 40, 50]
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Use map() to apply the conversion function to each element in the list
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))
fahrenheit_temps
```

```python
##10##  Create a Python program that uses `filter()` to remove all the vowels from a given string
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Function to remove vowels from a string
def remove_vowels(input_string):
    # Use filter() to remove vowels from the string
    return ''.join(filter(is_not_vowel, input_string))
remove_vowels("Hello, World!")
```

```python
orders= [[1, 15.5, 4],   # Order 1: 15.5€ per item, 4 items
    [2, 25.0, 3],   # Order 2: 25.0€ per item, 3 items
    [3, 50.0, 1],   # Order 3: 50.0€ per item, 1 item
    [4, 120.0, 2],  # Order 4: 120.0€ per item, 2 items
        ]

def calculate_total(order):
    order_number, price, quantity = order
    total = price * quantity
    # Add 10€ if the total is less than 100,00€
    if total < 10000:
        total += 10
    return (order_number, total)

# Use map() to apply the calculation to each order in the list
result = list(map(lambda order: calculate_total(order), orders))

result
```
